```python
def myreduce(fun,iterObj,initial=None):
    itr = iter(iterObj)
    if initial is None:
        a = next(itr)
    else:
        a = initial
    for b in itr:
        a=fun(a,b)
    return a
```

```python
myreduce(lambda a,b:a+b,[1,2,3,4,5])
```

```python
def add(a,b):
    return a+b
```

```python
r1 = myreduce(add,[1,2,3,4,5])
r1
```

```python
r2 = myreduce(add,[1,2,3,4,5],10)
r2
```

```python
sals = [1200,1500,2300,1800,1800,4500,3400,1900,5400,5300,4500]
```

```python
#add 10% tax for salries above 2000
#find total tax collected from employees
```

```python
def findSal(sal):
    if sal>2000:
        return True
    else:
        return False
```

```python
for sal in filter(findSal,sals):
    print(sal,end=" ")
```

```python
def tax(sal):
    return sal*0.20
```

```python
for taxSal in map(tax,filter(findSal,sals)):
    print(taxSal,end=" ")
```

```python
def sumOfTax(a,b):
    return a+b
```

```python
totalTax = reduce(sumOfTax,map(tax,filter(findSal,sals)))
totalTax
```
