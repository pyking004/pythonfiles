```python
#This is closure function
#if any function returns another function reference that is called closure function
#a closure == inner function + if any other preserved variables
def f1():
    def f2():
        return "Good Morning"
    return f2
```

```python
# Greet function has one preserved variable
def f1(msg):
    def greet(name):
        return msg +" "+ name
    return greet
```

```python
gm = f1("Good Morning")
```

```python
gm("Ravi")
```

```python
gm("Hari")
```

```python
gv = f1("Good Evening")
```

```python
gv("Ravi")
```

```python
#closure function
def discount(d):
    def sales(p):
        return p-(p/100*d)
    return sales
```

```python
daily_sales = discount(10)
```

```python
daily_sales(1000)
```

```python
weekly_sales = discount(20)
```

```python
weekly_sales(1000)
```

```python
def login():
    print("login success")
```

```python
#here f1 function is closure function
def f1(fun):
    def f2():
        print("***********Code before fun function **************")
        fun()
        print("###########Code after fun function ###############")
    return f2
```

```python
obj1 = f1(login)
```

```python
type(obj1)
```

```python
obj1()
```

```python
obj2 = f1(logout)
```

```python
obj2()
```

```python
login()
```

```python
# login2 = f1(login2)
# decorators  -> here @ is decorator
#here f1 must be closure function
@f1
def login2():
    print("User Login successfully!!")
```

```python
login2()
```

```python
def login3():
    print("Login Done")
```

```python
login3 = f1(login3)
```
