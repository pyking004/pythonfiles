```python
emp_sals = [2500,4500,3300,2800,3200]
```

```python
def gross_sal(sal):
    return (sal*0.4)+(sal*0.3)+(sal*0.2)+sal
```

```python
gross_sal(1000)
```

```python
grossSals = map(gross_sal,emp_sals)
```

```python
for i in grossSals:
    print(i)
```

```python
words =['python','welcome','java','good','morning']
```

```python
def vcnt(word):
    cnt=0
    vowels = ['a','e','i','o','u']
    for ch in word:
        if ch in vowels:
            cnt+=1
    return word,cnt
```

```python
vcnt("welcome")
```

```python
for i in map(vcnt,words):
    print(*i)
```
