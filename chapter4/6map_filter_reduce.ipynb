```python
def ov(ch):
    if ch in ['a','e','i','o','u']:
        return True
    else:
        return False
```

```python
for i in filter(ov,"welcome"):
    print(i,end=", ")
```

```python
def owcnt(s):
    cnt=0
    for i in s:
        if i in ['a','e','i','o','u']:
            cnt+=1
    return cnt
```

```python
owcnt("welcome")
```

```python
owcnt("Good Morning")
```

```python
bag_of_words = ['python','java','welcome','javascript','react','django']
```

```python
def owcnt(s):
    cnt=0
    for i in s:
        if i in ['a','e','i','o','u']:
            cnt+=1
    return s,cnt
```

```python
for i in map(owcnt,bag_of_words):
    word,cnt = i
    print(word,cnt)
```

```python
for i in map(owcnt,bag_of_words):
    print(*i)
```

```python
def reduce123(function,iterable,initial=None):
    it = iter(iterable)
    if initial is None:
        value = next(it)
    else:
        value = initial
    for element in it:
        value = function(value, element)
    return value
```

```python
reduce123(add,[1,2,3,4,5])
```

```python
reduce123(add,[1,2,3,4,5],10)
```
