```python
def vcnt(s):
    v = ['a','e','i','o','u']
    cnt = 0
    for i in s:
        if i in v:
            cnt+=1
    return cnt
```

```python
vcnt('welcome')
```

```python
bag_of_words = ['welcome','good','morning','python']
```

```python
for i in map(vcnt,bag_of_words):
    print(i)
```

```python
def vcnt(s):
    v = ['a','e','i','o','u']
    cnt = 0
    for i in s:
        if i in v:
            cnt+=1
    return s,cnt
```

```python
vcnt("welcome")
```

```python
for i in map(vcnt,bag_of_words):
    print(i)
```

```python
for word,cnt in map(vcnt,bag_of_words):
    print(word +" "+str(cnt))
```

```python
bag_of_words = ['welcome','good','morning','python','ramesh','amithkumar','kirankumar','varma']
```

```python
def vcnt2(s):
    v = ['a','e','i','o','u']
    cnt = 0
    for i in s:
        if i in v:
            cnt+=1
    return cnt>=3
```

```python
for i in filter(vcnt2,bag_of_words):
    print(i)
```

```python
for i in map(vcnt,filter(vcnt2,bag_of_words)):
    print(i)
```
