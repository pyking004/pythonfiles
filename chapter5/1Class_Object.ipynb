```python
class Car:
    tyres = 4   #class attribute
    def __init__(self,color):
        self.color = color  # dynamic object attribute
        print("{} color car created".format(self.color))
    def start(self):
        return "car started"
    def stop(self):
        return "car stopped"
    def changeColor(self,color):
        self.color = color
        return "Car color changed to {} color ".format(self.color)
```

```python
car1 = Car("Red")
```

```python
car1.color
```

```python
car2 = Car("Blue")
```

```python
car2.color
```

```python
car1.changeColor("Yellow")
```
