```python
class x:
    a = 10
    b = 20
    
    def m1(self):
        self.c = 300
        print(id(self))
        return "m1 of class x"
        
    @classmethod
    def m2(cls):
        print(id(cls))
        return "m2 of class x"

    @staticmethod
    def m3(a):
        x.a = a
        return "m3 of class x",x.a
```
