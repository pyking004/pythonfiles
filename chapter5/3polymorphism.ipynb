```python
#method overloading
class x:
    a = 100
    b = 200
    
    @Overload
    @signature()
    def m1(self):
        return "first m1"

    @m1.overload
    @signature('int')
    def m1(self,a):
        return "second m1"

    @m1.overload
    @signature('str')
    def m1(self,a):
        return "third m1"

    @m1.overload
    @signature('int','str')
    def m1(self,a,b):
        return "fourth m1"

    @m1.overload
    @signature('str','int')
    def m1(self,a,b):
        return "fifth m1"
```

```python
xobj1 = x()
```

```python
xobj1.m1()
```

```python
xobj1.m1(100)
```

```python
xobj1.m1("welcome")
```

```python
xobj1.m1(100,"python")
```

```python
xobj1.m1("Ravi",1001)
```

```python
class c1:
    a = 100
    def grossSal(self,sal):
        return (sal*0.4)+(sal*0.3)+(sal*0.25)+sal
```

```python
class c2(c1):
    b = 200
    def grossSal(self,basesal,bonus=0):
        bonus = basesal*bonus
        return super().grossSal(basesal)+bonus
```

```python
c2obj1 = c2()
```

```python
c2obj1.grossSal(1000)
```

```python
c2obj1.grossSal(2000)
```

```python
c2obj1.grossSal(2000,.2)
```

```python
c2obj1.grossSal(2000,.3)
```
