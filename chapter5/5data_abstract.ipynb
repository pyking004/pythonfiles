```python
class Vehicle(ABC):
    engine = 1
    brake = 1
    
    @abstractmethod
    def start(self):
        return "started"
    @abstractmethod
    def stop(self):
        return "stopped"
```

```python
class Car(Vehicle):
    tyres = 4
    @abstractmethod
    def start(self):
        return "Car "+super().start()
    @abstractmethod
    def stop(self):
        return "Car "+super().stop()
```
