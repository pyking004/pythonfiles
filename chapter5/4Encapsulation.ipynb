```python
Encapsulation:
    It is object-Oriented Programing pricipal 
    where data(variables) and methods(functions) that operate on that data are bundled inside a class
why Encapsulation?
    * protect data from accidental modification
    * control how data is accessed  or modified
    * improve maintainability and security
```

```python
#Encapsulation   ==> Data hiding
public      ==>  a   
protected   ==>  _a
private     ==>  __a
```

```python
class x:
    a = 100
    _b = 200
    __c = 300
    def m1(self):
        return "m1 of class x"
    def _m2(self):
        return "m2 of class x"
    def __m3(self):
        return "m3 of class x"
```

```python
class Bank2:
    def __init__(self,bal):
        self.__bal = bal  # private variable

    def deposit(self,amount):
        if amount > 0:
            self.__bal += amount

    def withdraw(self,amount):
        if 0 < amount <= self.__bal:
            self.__bal -= amount
        else:
            return "Unsufficient balance"

    def getBal(self):
        return self.__bal
```

```python
account1 = Bank2(1000)
```

```python
#here you can not access directly private attributes
account1.__bal = -500
```

```python
account1.getBal()
```

```python
account1.withdraw(500)
```

```python
account1.getBal()
```
