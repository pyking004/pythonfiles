```python
#single inheritance ==> one class  inherited from only one class
#super or base or parent
class c1:
    a = 100
    def m1(self):
        return "m1 of class c1"
```

```python
#sub or derived or child
class c2(c1):
    b = 200
    def m2(self):
        return "m2 of class c2"
```

```python
c2obj1 = c2()
```

```python
print(c2obj1.a)
print(c2obj1.b)
print(c2obj1.m1())
print(c2obj1.m2())
```
